Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?

Answer 1
Dimensionality Reduction is the process of reducing the number of input features while keeping important information intact.
It transforms high-dimensional data into a lower-dimensional space.

Why it is important:

Reduces computational cost (faster training).

Removes noise, improving model accuracy.

Prevents overfitting by simplifying data.

Improves visualization (e.g., converting 100-dimensional data to 2D/3D).

Question 2: Name and briefly describe three common dimensionality reduction
techniques..
Answer 2: Three Dimensionality Reduction Techniques

PCA (Principal Component Analysis):
Converts features into new components that capture maximum variance.

t-SNE (t-Distributed Stochastic Neighbor Embedding):
Used for visualizing high-dimensional data in 2D/3D; preserves local structure.

LDA (Linear Discriminant Analysis):
Supervised technique that separates classes by maximizing class separation.

Question 4: Explain the concept of anomaly detection and its significance.

Answer 4:  Anomaly Detection and its Significance

Anomaly Detection refers to identifying unusual patterns or data points that deviate from normal behavior.

Significance:

Detects fraud in banking.

Identifies network intrusions or cyber-attacks.

Helps detect machine failures in industries.

Useful for health monitoring and time-series issues.

Question 5: List and briefly describe three types of anomaly detection techniques.
Answer 5: Three Types of Anomaly Detection Techniques

Statistical Methods:
Detect anomalies based on probability distributions (e.g., Z-score).

Machine Learning Methods:
Algorithms like Isolation Forest, One-Class SVM learn normal patterns.

Deep Learning Methods:
Autoencoders, LSTM networks detect anomalies in complex data.

Question 6: What is time series analysis? Mention two key components of time series data.

Answer 6: What is Time Series Analysis? Two Key Components

Time Series Analysis deals with data collected over time to identify trends, patterns, and forecasts.

Two key components:

Trend: Long-term increase or decrease in data.

Seasonality: Repeating patterns occurring at fixed intervals.

Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Answer 7: Difference Between Seasonality and Cyclic Behaviour
Seasonality	Cyclic Behaviour
Occurs at fixed, known intervals (daily, monthly, yearly).	Occurs at irregular intervals.
Driven by weather, festivals, calendar effects.	Driven by business cycles, economy, or external events.
Predictable and repeatable.	Hard to predict.


```python
#Python Code for K-Means Clustering
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Create sample dataset
X, y = make_blobs(n_samples=200, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("K-Means Clustering Output")
plt.show()
```

```python
#What is inheritance in OOP? Provide a simple example in Python.
#Inheritance allows one class (child) to acquire attributes & methods from another class (parent).
class Animal:
    def speak(self):
        return "Animal makes sound"

class Dog(Animal):
    def bark(self):
        return "Dog barks"

d = Dog()
print(d.speak())  # inherited
print(d.bark())
```

Question 10: How can time series analysis be used for anomaly detection?
Answer 10: How Time Series Analysis Helps in Anomaly Detection

Time series analysis helps identify anomalies by understanding normal patterns of the data over time.

How it helps:

Detect sudden spikes/drops using moving averages.

Use forecasting models (ARIMA, LSTM): if actual value deviates from predicted, it is an anomaly.

Seasonal decomposition helps detect unusual seasonal behavior.

Control charts identify values that exceed statistical limits.
