1. What is Python and why is it popular?

Python is a high-level, interpreted programming language known for its simple syntax and readability. It is popular because it is easy to learn, supports multiple programming paradigms, and has large libraries for data science, web development, automation, and AI.





2. What is an interpreter in Python?

An interpreter is a program that executes Python code line by line without compiling the entire program beforehand. It converts Python instructions into machine code during execution.

3. What are pre-defined keywords in Python?

Pre-defined keywords are reserved words in Python that have special meanings and cannot be used as variable names.

Examples:
if, else, while, for, class, def, True, False.

4. Can keywords be used as variable names?

No. Python keywords cannot be used as variable names because they are reserved for specific functions in the language.

5. What is mutability in Python?

Mutability refers to whether an object can be changed after it is created.

Mutable objects → can change (list, dictionary, set)

Immutable objects → cannot change (tuple, string, integer)

6. Why are lists mutable but tuples immutable?

Lists are mutable because their elements can be modified after creation. Tuples are immutable, meaning their values cannot be changed once defined

7. Difference between == and is

== compares values

is compares memory locations
Example
a = [1,2]
b = [1,2]

print(a == b)  # True
print(a is b)  # False

13. Difference between for and while loop
For Loop	While Loop
Used when iterations are known	Used when condition-based
Iterates over sequence	Runs until condition becomes false

14. When is a while loop more suitable?

A while loop is suitable when the number of iterations is unknown and depends on a condition.

Example:

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop ?

Here is a simple Python program to calculate the factorial of a number using a while loop:

num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial = factorial * i
    i = i + 1

print("Factorial of", num, "is", factorial)

```python
#14. Write a program to reverse a string using a while loop.
text = input("Enter string: ")
i = len(text)-1

while i >= 0:
    print(text[i], end="")
    i -= 1
```
